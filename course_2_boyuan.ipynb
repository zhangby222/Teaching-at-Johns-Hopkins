# make it better by scraping all pages 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def url_generator(page):
    url = "https://english.elpais.com/news/politics/{}".format(page)
    return url
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
df = pd.DataFrame(columns=['title', 'link', 'author', 'place', 'datetime', 'description'])
```

```python
# put all codes together
def processor():
    global df
    title_soup = soup.find_all('h2', class_="c_h headline | color_gray_ultra_dark font_secondary width_full headline_md c_h__md")
    title_list = [i.text for i in title_soup]

    link_soup = title_soup.copy()
    link_list = [base + i.find('a').get('href') for i in link_soup]

    sideline_soup = soup.find_all('div', class_='c_b byline | uppercase color_gray_ultra_dark margin_bottom width_full')
    author_list = [i.find(class_="false").text for i in sideline_soup]
    place_list = []
    for i in sideline_soup:
        try:
            place = i.find(class_="capitalize").text
        except AttributeError:
            place = ""
        place_list.append(place)
    datetime_list = [i.find('time').get("datetime") for i in sideline_soup]
    description_soup = soup.find_all(class_="c_d description | color_gray_medium block width_full false false")
    description_list = [i.text for i in description_soup]

    df.title = title_list
    df.link = link_list
    df.author = author_list
    df.place = place_list
    df.datetime = datetime_list
    df.description = description_list
    return df



```

```python
df_total = pd.DataFrame(columns=['title', 'link', 'author', 'place', 'datetime', 'description'])
```

```python
for pageNum in range(3):
    url = url_generator(pageNum)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    processor()
    df_total = df_total.append(df)
    
    
```
