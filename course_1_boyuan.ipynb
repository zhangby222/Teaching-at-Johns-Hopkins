# 4. Functions, Modules

```python
step = 1
walk = 0
walk_total = 0
while step < 100:
    walk = np.random.rand()
    walk_total += walk
    # print("step{}, walked{}, walked total {}".format(step, walk, walk_total))
    step += 1

walk_avg = walk_total / step

print("step is",step)

print("average is", walk_avg)
```

```python
def walk_simulator(steps):
    step = 0
    walk = 0
    walk_total = 0
    while step < steps:
        walk = np.random.rand()
        walk_total += walk
        # print("step{}, walked{}, walked total {}".format(step, walk, walk_total))
        step += 1

    walk_avg = walk_total / step
    # print("step is", step)
    # print("average is", walk_avg)
    return walk_avg
```

```python
walk_simulator(1000)
```

```python
walk_simulator(1000000)
```

```python
def walk_sampler(times):
    step_list =[]
    walk_avg_list =[]
    for steps in range(times):
        steps += 1
        walk_avg = walk_simulator(steps)
        step_list.append(steps)
        walk_avg_list.append(walk_avg)
    return step_list, walk_avg_list
```

```python
step_list, walk_avg_list = walk_sampler(10000)
```

```python
'''
matplotlib
seaborn
'''
```

```python
import plotly.express as px
def walk_plotter(step_list, walk_avg_list):
    fig = px.scatter(x=step_list, y=walk_avg_list, labels={'x':'steps', 'y':'avg walks'})
    return fig
```

```python
walk_plotter(step_list, walk_avg_list)
```
